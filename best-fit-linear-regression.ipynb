```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

# Input data
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5, 7])
```

```python

# Create and train the model
model = LinearRegression()
model.fit(x, y)
```

```python

# Get slope and intercept
slope = model.coef_[0]
intercept = model.intercept_
```

```python

# Predict y values
y_pred = model.predict(x)
```

```python

# Print equation
print("Slope:", slope)
print("Intercept:", intercept)
print(f"Best fit line: y = {slope:.2f}x + {intercept:.2f}")

# Plot
plt.scatter(x, y, label="Actual data")
plt.plot(x, y_pred, label="Best fit line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression Best Fit Line")
plt.legend()
plt.show()
```

```python
# Step 1: Unseen data (new input values)
X_unseen = np.array([ [7], [8],[9]])  # Unseen input values

# Step 2: Use the trained model to predict the output for unseen data
y_unseen_pred = model.predict(X_unseen)

# Step 3: Print the predictions for the unseen data
print(f"Predictions for unseen data {X_unseen.flatten()}: {y_unseen_pred}")

# Optional: Visualize the unseen data predictions along with the original data and fitted line
plt.scatter(x, y, color='blue', label='Actual Data')  # Original data points
plt.plot(x, y_pred, color='red', label='Fitted Line')  # Fitted line
plt.scatter(X_unseen, y_unseen_pred, color='green', label='Unseen Data Predictions')  # Unseen data predictions
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: y = x (Testing with Unseen Data)')
plt.legend()
plt.show()
```

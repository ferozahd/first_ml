```python
# Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python

# Step 2: Generate data where y = x
X = np.array([[1], [2], [3], [4], [5]])  # Feature values (X)
y = np.array([1, 2, 3, 4, 5])  # Target values (y)
```

```python
# Step 3: Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)
```

```python

# Step 4: Make predictions
y_pred = model.predict(X)
```

```python

# Visualize the results
plt.scatter(X, y, color='blue', label='Actual Data')  # Original data points
plt.plot(X, y_pred, color='red', label='Fitted Line')  # Fitted line
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: y = x')
plt.legend()
plt.show()

# Print model parameters
print(f"Intercept: {model.intercept_}")
print(f"Slope: {model.coef_}")
```

```python
# Step 1: Unseen data (new input values)
X_unseen = np.array([[6], [7], [8]])  # Unseen input values

# Step 2: Use the trained model to predict the output for unseen data
y_unseen_pred = model.predict(X_unseen)

# Step 3: Print the predictions for the unseen data
print(f"Predictions for unseen data {X_unseen.flatten()}: {y_unseen_pred}")

# Optional: Visualize the unseen data predictions along with the original data and fitted line
plt.scatter(X, y, color='blue', label='Actual Data')  # Original data points
plt.plot(X, y_pred, color='red', label='Fitted Line')  # Fitted line
plt.scatter(X_unseen, y_unseen_pred, color='green', label='Unseen Data Predictions')  # Unseen data predictions
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression: y = x (Testing with Unseen Data)')
plt.legend()
plt.show()
```

1. What is `learning rate` ?
   1. The learning rate controls how big a step the model takes in the direction of the gradient during each update.
2. What is `epochs` ?
   1. One complete pass through all training samples
3. what is `weights` ?
4. what is `bias` ?
   1. Bias is the inteceptor of the streight line
5. what is `x_mean` ?
   1. Average of each features
6. what is `x_std` ?
   1. x_std stores the standard deviation of each feature.



$m = \frac{\sum (x_i-\bar{x})(y_i-\bar{y})}{\sum (x_i-\bar{x})^2}$

$b = \bar{y} - m\bar{x}$

```


class SimpleLinearRegression:
    def __init__(self):
        self.m = None   # slope
        self.b = None   # intercept

    def fit(self, X, y):

        # Convert to 1D arrays
        X = np.array(X).flatten()
        y = np.array(y).flatten()

        # Mean of x and y
        x_mean = np.mean(X)
        y_mean = np.mean(y)

        # Apply mathematical formulas directly

        numerator = np.sum((X - x_mean) * (y - y_mean))

        denominator = np.sum((X - x_mean) ** 2)

        # Slope
        self.m = numerator / denominator

        # Intercept
        self.b = y_mean - self.m * x_mean

    def predict(self, X):

        X = np.array(X).flatten()

        return self.m * X + self.b


```

```


class LinearRegressionGradientDescent:
    def __init__(self, lr=0.001, epochs=10000):
        self.lr = lr
        self.epochs = epochs
        self.weights = None
        self.bias = 0
        self.x_mean = None
        self.x_std = None

    def fit(self, X, y):
        n_samples, n_features = X.shape

        self.x_mean = np.mean(X, axis=0)
        self.x_std = np.std(X, axis=0)

        X = (X - self.x_mean) / self.x_std

        self.weights = np.zeros(n_features)
        self.bias = 0

        for _ in range(self.epochs):
            y_pred = np.dot(X, self.weights) + self.bias

            error = y_pred - y 

            dw = (1 / n_samples) * np.dot(X.T, error)
            db = (1 / n_samples) * np.sum(error)

            self.weights -= self.lr * dw
            self.bias -= self.lr * db

    def predict(self, X):
        X = (X - self.x_mean) / self.x_std
        return np.dot(X, self.weights) + self.bias
```

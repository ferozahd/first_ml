```
import numpy as np
import pandas as pd
import tensorflow as tf
import seaborn as sns
import matplotlib.pyplot as plt
```

```

# -----------------------------
# 1. Custom simple data
# -----------------------------
# X = Study hours
# y = Exam marks

data = {
    "Hours": [1, 2, 3, 4, 5, 6, 7, 8],
    "Marks": [15, 25, 32, 42, 50, 58, 68, 78]
}

df = pd.DataFrame(data)

print(df)
```

```

# -----------------------------
# 2. Prepare X and y
# -----------------------------
X = df[["Hours"]].values.astype(np.float32)
y = df["Marks"].values.astype(np.float32)
print(X)
print(y)
```

```

# -----------------------------
# 3. Create TensorFlow Linear Regression model
# -----------------------------
model = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(1)
])

# -----------------------------
# 4. Compile model
# -----------------------------
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.05),
    loss="mse"
)
```

```

# -----------------------------
# 5. Train model
# -----------------------------
model.fit(X, y, epochs=500, verbose=0)
```

```

# -----------------------------
# 6. Check learned weight and bias
# -----------------------------
weight, bias = model.layers[0].get_weights()

print("Weight:", weight[0][0])
print("Bias:", bias[0])
```

```

# -----------------------------
# 7. Predict unseen data
# -----------------------------
unseen_hours = np.array([[9], [10], [2.5]], dtype=np.float32)
```

```

predicted_marks = model.predict(unseen_hours)

print("\nUnseen Data Prediction:")
for hour, mark in zip(unseen_hours, predicted_marks):
    print(f"Study Hours: {hour[0]} -> Predicted Marks: {mark[0]:.2f}")
```

```
# Best fit line
x_line = np.linspace(1, 10, 100).astype(np.float32)
x_line[0]
```

```
# Best fit line
x_line = np.linspace(1, 10, 100).reshape(-1, 1).astype(np.float32)
y_line = model.predict(x_line)
x_line.shape
```

```

# -----------------------------
# 8. Draw scatter plot and best fit line
# -----------------------------
plt.figure(figsize=(8, 5))

# Actual data points
sns.scatterplot(x=df["Hours"], y=df["Marks"], s=100, label="Actual Data")



plt.plot(x_line, y_line, label="Best Fit Line")

# Unseen predicted points
sns.scatterplot(
    x=unseen_hours.flatten(),
    y=predicted_marks.flatten(),
    s=120,
    marker="X",
    label="Unseen Predictions"
)

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression using TensorFlow")
plt.legend()
plt.grid(True)
plt.show()
```

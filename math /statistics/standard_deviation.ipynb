# Standard Deviation

The **standard deviation** measures how spread out the values are from the mean.

### Equation:

$$
\sigma = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (x_i - \mu)^2}
$$

Where:  
- $\sigma$ = standard deviation  
- $x_i$ = each value in the dataset  
- $\mu$ = mean of the dataset  
- $n$ = total number of values

```python
import numpy as np 
import math 
X = np.array([2, 4, 4, 4, 5, 5, 7, 9])
```

```python
std_X = np.std(X)
print("Standard Deviation:", std_X)
```

## Step by step calculation 

```python
X_mean =np.mean(X)
print(f'Mean value is {X_mean}')
mean_deviation=X-X_mean
print(f'Mean deviation is {mean_deviation}')
squre_deviation=mean_deviation ** 2
print(f'Squre deviation is {squre_deviation}')


average_squre_deviation=np.mean(squre_deviation)
print(f'Average squre deviation is {average_squre_deviation}')

std_deviation=math.sqrt(average_squre_deviation)
print(f'Standard deviation is {std_deviation}')
```

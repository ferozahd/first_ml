# Understanding Mean in Python

This document explains **mean** and **standard deviation** using a custom dataset and `numpy`.

---

## 1. Sample Data

We have a small dataset:

```python
import numpy as np

X = np.array([2, 4, 4, 4, 5, 5, 7, 9])
```

The mean is $ \mu = \frac{1}{n} \sum_{i=1}^{n} x_i $.

````python
import numpy as np

# Custom dataset
X = np.array([2, 4, 4, 4, 5, 5, 7, 9])

mean_x =np.mean(X)
print(f'mean value {mean_x}')
````

## Custom python code to find mean value

````python
total =0
for each in X: 
    total += each 
print(f'Total is {total}')
mean_x=total/X.size
print(f'Mean is {mean_x}')
````

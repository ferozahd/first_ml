```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python

# Create a sample DataFrame
data = {'X1': [1, 2, 3, 4, 5], 'X2': [4, 5, 6, 7, 8], 'y': [2, 4, 6, 8, 10]}
df = pd.DataFrame(data)
```

```python

# Features and target variable
X = df[['X1', 'X2']]  # Two input columns
y = df['y']  # Output column
```

```python

# Apply PolynomialFeatures (degree = 5 for overfitting)
# poly = PolynomialFeatures(degree=5)
# X_poly = poly.fit_transform(X)
# X_poly
```

```python

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

```python

# Predict values
y_pred = model.predict(X)
```

```python

# Plot to visualize overfitting
plt.scatter(X['X1'], y, color='blue', label='Actual data')
# plt.scatter(X['X1'], X['X2'], color='green', label='Actual data')
plt.plot(X['X1'], y_pred, color='red', label='Overfitted Model', linewidth=2)
plt.xlabel('X1')
plt.ylabel('y')
plt.legend()
plt.title('Overfitting Example')
plt.show()

# Print the model coefficients and intercept
print("Model coefficients:", model.coef_)
print("Model intercept:", model.intercept_)
```

```python
X_unseen = pd.DataFrame({'X1':[6,7,8,9,10,11,12],'X2':[9,10,11,12,13,14,15]})
# X_unseen_poly=poly.fit_transform(X_unseen)
```

```python


# Step 2: Use the trained model to predict the output for unseen data
y_unseen_pred = model.predict(X_unseen)


# Plot to visualize overfitting
plt.scatter(X['X1'], y, color='blue', label='Actual data')
plt.scatter(X_unseen['X1'], y_unseen_pred, color='green', label='Actual data')
plt.plot(X['X1'], y_pred, color='red', label='Overfitted Model', linewidth=2)
plt.xlabel('X1')
plt.ylabel('y')
plt.legend()
plt.title('Overfitting Example')
plt.show()

# Print the model coefficients and intercept
print("Model coefficients:", model.coef_)
print("Model intercept:", model.intercept_)
```

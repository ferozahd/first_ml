```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
import matplotlib.pyplot as plt
```

```python

# -----------------------------------
# 1. Create simple custom data
# -----------------------------------
data = {
    "hours_studied":    [1, 2, 2, 3, 4, 5, 6, 7, 1, 3, 6, 8],
    "attendance":       [40, 50, 55, 60, 65, 70, 80, 90, 35, 75, 85, 95],
    "assignments_done": [0, 1, 1, 2, 2, 3, 3, 4, 0, 2, 4, 5],
    "result":           ["Fail", "Fail", "Fail", "Pass", "Pass", "Pass", "Pass", "Pass", "Fail", "Pass", "Pass", "Pass"]
}

df = pd.DataFrame(data)

print("Dataset:")
print(df)
```

```python

# -----------------------------------
# 2. Select input and output
# -----------------------------------
X = df[["hours_studied", "attendance", "assignments_done"]]
y = df["result"]
```

```python

# -----------------------------------
# 3. Create and train the Decision Tree model
# -----------------------------------
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X, y)
```

```python

# -----------------------------------
# 4. Predict a new student's result
# -----------------------------------
new_student = pd.DataFrame({
    "hours_studied": [2],
    "attendance": [58],
    "assignments_done": [1]
})

prediction = model.predict(new_student)

print("\nNew student data:")
print(new_student)
print("\nPredicted result:", prediction[0])
```

```python

# -----------------------------------
# 5. Show the tree rules in text form
# -----------------------------------
rules = export_text(model, feature_names=list(X.columns))
print("\nDecision Tree Rules:\n")
print(rules)
```

```python

# -----------------------------------
# 6. Draw the tree
# -----------------------------------
plt.figure(figsize=(12, 7))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=model.classes_,
    filled=True
)
plt.title("Decision Tree Example")
plt.show()
```

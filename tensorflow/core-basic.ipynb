````python
from typing import Literal
# This method use for checking tensor type either its @return Variable type or @return Constant type
# @params tensor identifier 
# @ return Literal Variable or Constant
def tensor_dtype(x )->Literal["Variable" , "Constant"]:
    if isinstance(x, tf.Variable):
        return "Variable"
    else:
        return "Constant"
````

````python
# Creating constant tensor 
const_tensor =tf.constant([1,2,3,4,5])
print(const_tensor)
print(type(const_tensor))
print(tensor_dtype(const_tensor))
````

````python
variable_tensor =tf.Variable([1,2,3,4,5],dtype=float)
print(variable_tensor)
print(type(variable_tensor))
print(tensor_dtype(variable_tensor))
````

## Tensor Rank 
Tensor rank actually refer the degree of tensors 
Rank 0
- Degree: 1 degree tensor
- Name: Scalar

Rank 1
- Degree: 2 degree tensor
- Name: Vector

Rank 2
- Degree: 3 degree tensor
- Name: Matrix

Rank 3
- Degree: 4 degree tensor
- Name: Cube / 3D Tensor

## Reduction operation 


````python

value1 = tf.constant([
    [1, 2, 3],
    [4, 5, 6]
],dtype=tf.float32)
# Reduction sum 
total_value1=tf.reduce_sum(value1)
# Inernal operation 
# Converting N-D to 0 D array and sum all them together 
print(f'Tensor reduction sum {total_value1.numpy()}')
# Internal operation 
# reducation_sum/ tensor size 
#21/6 =3.5
mean_value1= tf.reduce_mean(value1)
print(f'Tensor reduction mean {mean_value1.numpy()}')

max_value1 =tf.reduce_max(value1)
# Internal operation 
# Find largest value inside 0D tensor array 
print(f'Tensor reduction max {max_value1.numpy()}')

min_value1=tf.reduce_min(value1)
# Internal operation 
# Find smallest value inside 0D tensor array 
print(f'Tensor reduction mix {min_value1.numpy()}')
````

## Random Tensor 

````python
random_normal_tensor = tf.random.normal(shape=(2, 3))
print(random_normal_tensor.numpy())
````

````python
random_normal_tensor= tf.random.normal(
    shape=(2, 2),
    mean=10,
    stddev=2
)
print(random_normal_tensor)
````

````python
random_uniform_tensor = tf.random.uniform(
    shape=(2, 3),
    minval=0,
    maxval=10
)

print(random_uniform_tensor)
````

## File preprocessing 


```python
import os
import random
import shutil
from pathlib import Path

# =========================
# CONFIGURATION
# =========================

SOURCE_DIR = "data/cat_dog"   # your current folder containing all 25k images
OUTPUT_DIR = "data/cat_dog"

TRAIN_SPLIT = 0.70
VAL_SPLIT = 0.15
TEST_SPLIT = 0.15

random.seed(42)

# =========================
# CREATE OUTPUT DIRECTORIES
# =========================

splits = ["train", "validation", "test"]
classes = ["cats", "dogs"]

# for split in splits:
#     for cls in classes:
#         os.makedirs(os.path.join(OUTPUT_DIR, split, cls), exist_ok=True)

# =========================
# COLLECT FILES
# =========================

source_path = Path(SOURCE_DIR)

cat_files = list(source_path.glob("cat.*.jpg"))
dog_files = list(source_path.glob("dog.*.jpg"))

print(f"Found {len(cat_files)} cat images")
print(f"Found {len(dog_files)} dog images")
```

```python

# =========================
# SHUFFLE FILES
# =========================

random.shuffle(cat_files)
random.shuffle(dog_files)

# =========================
# SPLIT FUNCTION
# =========================

def split_files(files):

    total = len(files)

    train_end = int(total * TRAIN_SPLIT)
    val_end = train_end + int(total * VAL_SPLIT)

    train_files = files[:train_end]
    val_files = files[train_end:val_end]
    test_files = files[val_end:]

    return train_files, val_files, test_files

cat_train, cat_val, cat_test = split_files(cat_files)
dog_train, dog_val, dog_test = split_files(dog_files)

# =========================
# COPY FILES
# =========================

def copy_files(files, split_name, class_name):

    destination = Path(OUTPUT_DIR) / split_name / class_name

    for file in files:
        shutil.copy(file, destination / file.name)

# copy_files(cat_train, "train", "cats")
# copy_files(cat_val, "validation", "cats")
# copy_files(cat_test, "test", "cats")

# copy_files(dog_train, "train", "dogs")
# copy_files(dog_val, "validation", "dogs")
# copy_files(dog_test, "test", "dogs")

print("\nDataset split completed successfully.")
```

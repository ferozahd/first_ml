```python
import tensorflow as tf
```

```python

# Create data
x = tf.range(20, dtype=tf.float32)
y = x * 2
```

```python

# Create dataset
dataset = (
    tf.data.Dataset
    .from_tensor_slices((x, y))
    .shuffle(20)
    .batch(4)
)
```

```python
# Create model
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)
])

# Compile model
model.compile(
    optimizer='sgd',
    loss='mse'
)

# Train
model.fit(dataset, epochs=5)
```

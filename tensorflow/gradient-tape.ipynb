## GradientTape 
simplify the equation

$y=x^2$
Equeation in integration 

$ \frac{dy}{dx}=x^2 $
when `x=3` gradient should be

$6$

```python
import tensorflow as tf
```

```python
# Variable
x = tf.Variable(3.0)

# Record operations
with tf.GradientTape() as tape:
    y = x ** 2

# Compute derivative dy/dx
gradient = tape.gradient(y, x)

print("y =", y.numpy())
print("dy/dx =", gradient.numpy())
```

```python
# Data
x = tf.constant([1.0, 2.0, 3.0])
y_true = tf.constant([2.0, 4.0, 6.0])

# Trainable parameter
w = tf.Variable(1.0)

with tf.GradientTape() as tape:

    # Prediction
    y_pred = w * x

    # Loss
    loss = tf.reduce_mean((y_true - y_pred) ** 2)

# Gradient of loss w.r.t w
grad = tape.gradient(loss, w)

print("Loss:", loss.numpy())
print("Gradient:", grad.numpy())
```
